```python
#**Q1:** Using pandas, import the hotels.csv and show the number of observations and columns.
import pandas as pd
hotel= pd.read_csv('hotels.csv',index_col = 0)
#print(hotel.shape)
#print(hotel.head())
hotel.info()
```

```python
# **Q2:** Write a Python program to determine the number of hotel types and how many of each type.
hotel_type_counts = hotel['hotel_type'].value_counts()
print("Number of Hotel Types:", len(hotel_type_counts)) # using len function count and print number of hotels.
print("\nCount of Each Hotel Type:")
print(hotel_type_counts)
```

```python
# Q.3 Using a dataframe.method1().method2() syntax, count the number of missing values in each column.

missing_values=hotel.isna().sum()
print(missing_values)
```

```python
# Q4. a table showing the mean, standard deviation, skewness and kurtosis of all the continuous variables.

continuous_variables = []      # Initialize a list to store the names of continuous variables

for column in hotel.columns:
    if pd.api.types.is_numeric_dtype(hotel[column]):  # Check if the column is numeric (integer or float) 
        num_distinct_values = hotel[column].nunique() #Calculate the number of distinct values in the column
        if num_distinct_values > 10:# If the number of distinct values is greater than 10, consider it a continuous variable
            continuous_variables.append(column) # append values to continuous variable  

#print(continuous_variables)
hotel[continuous_variables].agg(['mean', 'std', 'skew', 'kurtosis']) # print table showing 4 moments of continuous variable
```

```python
# Kurtosis values before removing outliers.
from scipy.stats import  kurtosis 

kurtosis_before_outliers = hotel[continuous_variables].apply(kurtosis) # diplaying kurtosis values before handling outliers
kurtosis_before_outliers_list = kurtosis_before_outliers.tolist() # convert them to list
print(kurtosis_before_outliers)
```

```python
 #Q.5.Find and replace all continuous variable outliers with missing values
    
import numpy as np 

for variables in hotel[continuous_variables]:
    uppar_quartile = np.nanpercentile(hotel[variables],75) #define upaar limit
    lower_quartile = np.nanpercentile(hotel[variables],25) #define lowee limit
    iqr = uppar_quartile - lower_quartile
    
    uppar_whiskar = uppar_quartile + 1.5*iqr
    lower_whiskar = np.max(lower_quartile - 1.5*iqr,0)
    
    hotel[variables] = np.where(((hotel[variables] >= uppar_whiskar) | 
                                           (hotel[variables] <= lower_whiskar)) &
                                          (((uppar_whiskar - lower_whiskar) != 0)), np.nan , hotel[variables])   
               
hotel[continuous_variables].isnull().sum()
```

```python
# Find and replace missing  values.

hotel[continuous_variables].isnull().head(5)
```

```python
# Q6: Impute all continuous variable missing values with the means of those variables.

for variables in continuous_variables:
    hotel[variables].fillna(value = hotel[variables].mean(), inplace = True)
print(hotel[continuous_variables].head(10))  
```

```python
# find 4 moments after removing outleiers

import numpy as np
hotel[continuous_variables].agg(['mean', 'std', 'skew', 'kurtosis'])
```

```python
# Kurtosis values after Removing Outliers.

kurtosis_after_removing_outliers=hotel[continuous_variables].apply(kurtosis) # kurtosis values after removing outliers
kurtosis_after_removing_outliers_list = kurtosis_after_removing_outliers.tolist() # convert them to list
print(kurtosis_after_removing_outliers_list) # print
```

```python
#Q7: How has the kurtosis changed for each of the variables that had outliers removed and then imputed?
#Be sure that all the kurtosis values are less than those before outlier removal.

if kurtosis_after_removing_outliers_list[0:10] < kurtosis_before_outliers_list[0:10]: # comparing 2 kurtosis value 
    print("Result: Kurtosis is reduced after removing outliers")
else:   
    print("Result: Kurtosis is not reduced after removing outliers or remains the same")
```
